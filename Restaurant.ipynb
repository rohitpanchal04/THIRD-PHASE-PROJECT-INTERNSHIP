**Filling Null Values**

```python
# Filling Null Values
df['Cuisines'].fillna(df['Cuisines'].mode()[0],inplace=True)
df['Longitude'].replace(0.000000, df['Longitude'].mean(),inplace=True)
df['Latitude'].replace(0.000000,df['Latitude'].mean(),inplace=True)
```

```python
city_type=df['City'].value_counts(ascending=False)
city_lessthan200=city_type[city_type<200]
```

```python

def handlecity(value):
    if(value in city_lessthan200):
        return 'others'
    else:
        return value

df['City']=df['City'].apply(handlecity)
df['City'].value_counts()
```

```python
cuisines_type=df['Cuisines'].value_counts()
cuisines_lessthan300=cuisines_type[cuisines_type<300]
```

```python
def handlecuisines(value):
    if (value in cuisines_lessthan300):
        return 'others'
    else:
        return value

df['Cuisines']=df['Cuisines'].apply(handlecuisines)
df['Cuisines'].value_counts()
    
```

# PREDICTING - 'Average cost for Two'

```python
# Importing Machine Learning Library (Regression)
from sklearn.neighbors import KNeighborsRegressor  
from sklearn.tree import DecisionTreeRegressor 
from sklearn.linear_model import LinearRegression 
from sklearn.linear_model import Lasso,Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.metrics import f1_score
```

```python
# CHecking the Regression model which works better there
KNR=KNeighborsRegressor()
DTR=DecisionTreeRegressor()
LR=LinearRegression()
RFR=RandomForestRegressor()
ls=Lasso()
rd=Ridge()

model = [KNR,DTR,LR,RFR,ls,rd]
max_r2_score=0
for r_state in range(45,85):
    x_train, x_test, y_train,y_test=train_test_split(x,y,random_state=r_state, test_size=0.25)
    for i in model:
        i.fit(x_train, y_train)
        pre = i.predict(x_test)
        r2_sc=r2_score(y_test,pre)
        print('R2 Score correspond to random state',r_state,'is',r2_sc)
        if r2_sc>max_r2_score:
            max_r2_score=r2_sc
            final_state2=r_state
            final_model=i
print()
print('max r2 score correspond to random state',final_state2,'is',max_r2_score, 'and model is',final_model)
```

**BEST MODEL IS RANDOMFORESTCLASSIFIER AND R2 SCORE - 90%**
